### <font color="purple">**Question2**</font>
- Fill all the null values in the dataframe with '-1' and store the result to variable 'new_df'.

```python
#fill null values with -1 and store it to 'new_df'
new_df = df.fillna(-1)


#show the dataframe 'new_df'
new_df
```

### <font color="purple">**Question3**</font>
Fill the null values in respective columns like below:

    - apple(1kg): Mean value of that column
    - banana(1 dozen): Mean value of that column
    - grapes(1kg): Median value of that column
    - mango(1kg): Median value of that column
    - Water Melons(1): Fill null values with "Not Available".
    
Finally store the output to 'new_df' variable

```python
new_df = df.copy()

# Fill null values in 'apple(1kg)' column with the mean value of that column
new_df['apple(1kg)'].fillna(new_df['apple(1kg)'].mean(), inplace=True)

# Fill null values in 'banana(1 dozen)' column with the mean value of that column
new_df['banana(1 dozen)'].fillna(new_df['banana(1 dozen)'].mean(), inplace=True)

# Fill null values in 'grapes(1kg)' column with the median value of that column
new_df['grapes(1kg)'].fillna(new_df['grapes(1kg)'].median(), inplace=True)

# Fill null values in 'mango(1kg)' column with the median value of that column
new_df['mango(1kg)'].fillna(new_df['mango(1kg)'].median(), inplace=True)

# Fill null values in 'Water Melons(1)' column with "Not Available"
new_df['Water Melons(1)'].fillna("Not Available", inplace=True)

# Display the updated DataFrame
new_df
```
